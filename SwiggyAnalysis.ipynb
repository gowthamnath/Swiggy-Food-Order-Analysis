```python
# veg vs non-veg  ( 0 --> veg, 1 --> non-veg, 2 --> desert)
items.groupby('is_veg')['is_veg'].count()
```

```python
# create a new column chicken found (if chicken is found '0 or +ve value', not found '-1')
items['chicken_found'] = items['name'].str.find('Chicken')
items
```

```python
# to find chicken items only
items[items['chicken_found']!=-1]
```

```python
#create a new column paratha_found
items['paratha_found']= items['name'].str.find('Paratha')
items
```

```python
#to find onlyparatha items
items[items['paratha_found']!=-1]
```

```python
#items orders most time
items_new = items.groupby('name')['id'].count().reset_index()
items_new.sort_values(by=['id'],ascending = False)
```

```python
#orders dataset
#to find the distinct orders
orders['order_id'].unique()
```

```python
#to find the count of unique orders 
orders['order_id'].nunique()
```

```python
#to find the unique restaurant names
orders['restaurant_name'].unique()
```

```python
#restaurant and the number of orders
orders_new = orders.groupby('restaurant_name')['order_id'].nunique().reset_index()
orders_new = orders_new.rename(columns={'order_id':'number_of_orders'})
orders_new
```

```python
# to find the rank of the restaurant based on the number of orders
orders_new['rank'] = orders_new['number_of_orders'].rank(method = 'first' , ascending = False)
orders_new.sort_values(by=['rank'])
```

```python
#converting to date time and just getting the year from it
orders['order_time']= orders['order_time'].apply(pd.to_datetime)
orders['year'] = orders['order_time'].dt.year
orders
```

```python
#to get the year-wise revenue
Orders2 = orders.groupby('year')['order_total'].sum().reset_index()
Orders2
```

```python
# to get previous year revenue and percentage difference
Orders2['Previous_Year_revenue'] = Orders2['order_total'].shift(1)
Orders2
Orders2['Percentage_diff'] = (Orders2['order_total'] - Orders2['Previous_Year_revenue'])/Orders2['Previous_Year_revenue']*100
Orders2
```
